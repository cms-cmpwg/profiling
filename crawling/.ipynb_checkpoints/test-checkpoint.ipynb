```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
req=requests.get("https://jiwoong.web.cern.ch/jiwoong/cgi-bin/igprof-navigator/igprofCPU_")
html = req.text
soup = BeautifulSoup(html,'html.parser')
```

```python
columns=soup.select('tr > th')
```

```python
columnlist=[]
for column in columns:
    columnlist.append(column.text)
print(columnlist)
df=pd.DataFrame(columns=columnlist)
```

```python
contents = soup.select('tr')
dfcontent=[]
alldfcontents=[]

for content in contents:
    tds=content.find_all("td")
    for td in tds:
        dfcontent.append(td.text)
    alldfcontents.append(dfcontent)
    dfcontent=[]
df=pd.DataFrame(columns=columnlist, data=alldfcontents)
```
