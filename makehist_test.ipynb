```python
evt_list=[]
time_list=[]
VSIZE_list=[]
RSS_list=[]
column=[]

file1='SUM_CMSSW_10_6_0_pre3.txt'
file2='SUM_CMSSW_10_6_0_pre4.txt'

with open(file1) as f:
	for line in f:
		line=line.strip()
		#print(line)
		list_line1=line.split()
		
		nextline=next(f)
		nextline=nextline.strip()
		#print(nextline)
		list_line2=nextline.split()
		list_line= list_line1 + list_line2

		evt_list.append(list_line[10])
		time_list.append(list_line[12])
		VSIZE_list.append(list_line[4])
		RSS_list.append(list_line[7])
		print(list_line)

evt_list2=[]
time_list2=[]
VSIZE_list2=[]
RSS_list2=[]
column=[]

with open(file2) as f:
	for line in f:
		line=line.strip()
		#print(line)
		list_line1=line.split()
		
		nextline=next(f)
		nextline=nextline.strip()
		#print(nextline)
		list_line2=nextline.split()
		list_line= list_line1 + list_line2

		evt_list2.append(list_line[10])
		time_list2.append(list_line[12])
		VSIZE_list2.append(list_line[4])
		RSS_list2.append(list_line[7])
		print(list_line)
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('Agg')
```

```python
fig,axs = plt.subplots(2,3,figsize=(25,15))
axs[0,0].plot(evt_list,RSS_list,'o',color='r',alpha=0.7)
axs[0,0].plot(evt_list2,RSS_list2,'o')
axs[0,0].set_title('(RSS)Memory Profile',fontsize=20)
axs[0,0].set_xlim([0,10.1])
axs[0,0].set_xlabel('ith event',fontsize=15)
axs[0,0].set_ylabel('Memory(MB)',fontsize=15)
axs[0,0].legend(['CMSSW_10_6_0_pre3','CMSSW_10_6_0_pre4'],fontsize=15)
axs[0,0].xaxis.set_label_coords(0.96, -0.05) 


axs[0,1].plot(evt_list,VSIZE_list,'o',color='r',alpha=0.7)
axs[0,1].plot(evt_list2,VSIZE_list2,'o')
axs[0,1].set_title('(VSIZE)MEmory Profile',fontsize=20)
axs[0,1].set_xlim([0,10.1])
axs[0,1].set_xlabel('ith event',fontsize=15)
axs[0,1].set_ylabel('Memory(MB)',fontsize=15)
axs[0,1].xaxis.set_label_coords(0.96, -0.05)

axs[0,2].plot(evt_list,time_list,'o',color='r',alpha=0.7)
axs[0,2].plot(evt_list2,time_list2,'o')
axs[0,2].set_title('CPU Time Profile',fontsize=20)
axs[0,2].set_xlim([0,10.1])
axs[0,2].set_xlabel('ith event',fontsize=15)
axs[0,2].set_ylabel('time (seconds)',fontsize=15)
axs[0,2].xaxis.set_label_coords(0.96, -0.05)


n_bins=20

axs[1,0].hist(RSS_list,bins=n_bins,color='r',alpha=0.7)
axs[1,0].hist(RSS_list2,bins=n_bins,color='b',alpha=0.7)
axs[1,0].set_title('(RSS)Memory Profile',fontsize=20)
axs[1,0].set_ylim([0,11])
axs[1,0].set_ylabel('ith event',fontsize=15)
axs[1,0].set_xlabel('Memory(MB)',fontsize=15)
axs[1,0].xaxis.set_label_coords(0.96, -0.1)

axs[1,1].hist(VSIZE_list,bins=n_bins,color='r',alpha=0.7)
axs[1,1].hist(VSIZE_list2,bins=n_bins,color='b',alpha=0.7)
axs[1,1].set_title('(VSIZE)Memory Profile',fontsize=20)
axs[1,1].set_ylim([0,11])
axs[1,1].set_ylabel('ith event',fontsize=15)
axs[1,1].set_xlabel('Memory(MB)',fontsize=15)
axs[1,1].xaxis.set_label_coords(0.96, -0.05)

axs[1,2].hist(time_list,bins=n_bins,color='r',alpha=0.7)
axs[1,2].hist(time_list2,bins=n_bins,color='b',alpha=0.7)
axs[1,2].set_title('CPU Time Profile',fontsize=20)
axs[1,2].set_ylim([0,11])
axs[1,2].set_ylabel('ith event',fontsize=15)
axs[1,2].set_xlabel('time (seconds)',fontsize=15)
axs[1,2].xaxis.set_label_coords(0.9, -0.1)


for label in axs[1,0].xaxis.get_ticklabels():
    label.set_rotation(45)
for label in axs[1,2].xaxis.get_ticklabels():
    label.set_rotation(45)

plt.tight_layout()
plt.savefig("Summary.png")
```
